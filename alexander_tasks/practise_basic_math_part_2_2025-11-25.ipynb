Два користувачі: \
u1 = (50 переглядів, 10 лайків, 4 підписки) \
u2 = (70, 5, 1)



```python
import numpy as np

u1 = np.array([50, 10, 4])
u2 = np.array([70, 5, 1])

len_u1 = np.linalg.norm(u1)
len_u2 = np.linalg.norm(u2)


if len_u1 > len_u2:
    print("len_u1 winner")
else:
    print("len_u2 winner")    



# use linalg.norm
```

**Task 5.** Нормалізація поведінкових фіч перед кластеризацією


```python
import numpy as np

users = np.array([
    [10, 3, 1],
    [40, 2, 0],
    [20, 5, 2],
    [15, 1, 1],
    [60, 7, 3]
])

print(np.linalg.norm(users))
users_1 = np.linalg.norm(users)

print(np.linalg.norm(users) / users_1)

norm_ = np.linalg.norm(users, axis = 1, keepdims=True)

normalize_users = users / norm_

print(normalize_users)

# use linalg.norm
# users / by previos step
```

**Task 5.** Оцініть наскільки користувачу підходить товар

```python
# user data 
#u = (0.2, 0.8, 0.3, 0.1)

# product data
#p = (0.25, 0.75, 0.2, 0.15)


import numpy as np

u = np.array([0.2, 0.8, 0.3, 0.1])
m = np.array([0.25, 0.75, 0.2, 0.15])

score = np.dot(u, m) 
print(score)
```

```python
u = np.array([0.2, 0.8, 0.3, 0.1])
m = np.array([0.25, 0.75, 0.2, 0.15])

def cos_sim(u, m):
    return np.dot(u,m) / (np.linalg.norm(u) * np.linalg.norm(m))

print(cos_sim(u, m))
```

```python
# Cosine similarity показує, наскільки два вектори схожі за напрямом, ігноруючи їхній масштаб.
# Косинус кута між ними:
# - 1 — стрілки співпадають → повна схожість
# - 0 — стрілки перпендикулярні → немає зв’язку
# - (-1) — стрілки протилежні → повна несумісність
```
