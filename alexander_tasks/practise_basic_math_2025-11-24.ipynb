**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]

arr.sort() # sorted 

n = len(arr)

if n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print(median_) 

```

**Задача 2** \
Дані: [3, 3, 4, 100] \
Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100]

#task 1
temp = sorted(arr) #[::-1]

# 100, 1, 5, 100, 150, 10

n = len(arr)

if n % 2 == 1:
    median_ = temp[n // 2]
else:
    median_ = (temp[n // 2 - 1] + arr[n // 2]) / 2

avg_arr = sum(arr) / n
print(median_, avg_arr)    
```
